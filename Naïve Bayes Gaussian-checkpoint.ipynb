```python
df=pd.read_csv("UniversalBank.csv")
df.head(20)
```

### Dropping Unneccessary columns

```python
data = df.drop(["ID","ZIP Code"],axis=1)
```

### Creating Dummies for the categorical is present

```python
cat_cols = ["Family","Education","Personal Loan","Securities Account","CD Account","Online","CreditCard"]
data = pd.get_dummies(data,columns=cat_cols,drop_first=True,)
```

```python
data.head()
```

### Splitting the data into Training and Test

```python
X = data.copy().drop("Personal Loan_1",axis=1)
y = data["Personal Loan_1"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Algorithms

#### Na√Øve Bayes Gaussian algorithm

Naive Bayes algorithm considers the variable as an independent variable.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

### Linear Discriminant Algorithm

Linear Discriminant algorithm considers that your variables are Gaussian and have same mean.

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train,y_train)
predictions_lda = lda.predict(X_test)
print("Accuracy Score is:")
print(accuracy_score(y_test, predictions_lda))
```

### Summary:

Both Linear Discriminant Algorithm (LDA) and Na√Øve Bayes (NB) are linear classifiers and come under the category of Generative Models which estimates the posterior P(class,x).

LDA is closely related to NB in that both classifiers assume Gaussian within-class distributions. However, NB relies on a less flexible distributional model .An LDA classifier is Bayes-optimal (ignoring estimation error) if the distributions corresponding to the two classes are Gaussian and have equal covariance.

In our data set we can see that Naive bayes accuracy less than that of Linear Discriminant Analysis.

So, we can conclude the LDA gives a better accuracy than Na√Øve bayes for the given data set.
